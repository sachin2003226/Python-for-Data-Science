ProblemStatement:
Solve the following:

Given a matrix mat of size 3 x 3. Find the maximum numbers from each row, say N1, N2 and N3. Result in a matrix by adding:


import numpy as np
mat = np.array([[10,5,9],
                [2,20,6],
                [8,3,30]]).reshape(3,3)


N1 to the upper half elements of mat

N2 to the main diagonal elements of mat

N3 to the lower half elements of mat



Answer/Hint:
[[30 15 19]
[32 40 16]
[38 33 50]] Hint: Use tril_indices(), mask_indices(), diag_indices()

```python
import numpy as np

# Given matrix
mat = np.array([[10, 5, 9],
                [2, 20, 6],
                [8, 3, 30]])

# Step 1: Max of each row
N1 = np.max(mat[0])  # 10
N2 = np.max(mat[1])  # 20
N3 = np.max(mat[2])  # 30

# Step 2: Get indices
upper_indices = np.triu_indices(3, k=1)
lower_indices = np.tril_indices(3, k=-1)
diag_indices = np.diag_indices(3)

# Step 3: Add to matrix
result = mat.copy()
result[upper_indices] += N1
result[diag_indices] += N2
result[lower_indices] += N3

# Final result
print(result)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.data import shepp_logan_phantom
from skimage.transform import resize

# 1. Load the phantom image (grayscale float image from 0 to 1)
img = shepp_logan_phantom()

# Show the original image
plt.imshow(img, cmap='gray')
plt.title("Original Phantom Image")
plt.axis('off')
plt.show()

# 2. Black and white thresholding at 0.15
new_img = np.where(img <= 0.15, 0, 1)

# Show the thresholded image
plt.imshow(new_img, cmap='gray')
plt.title("Thresholded Image (0.15)")
plt.axis('off')
plt.show()

# 3. Mirror image using fliplr (flip left to right)
flipped_img = np.fliplr(new_img)

# Show the flipped image
plt.imshow(flipped_img, cmap='gray')
plt.title("Flipped Image (Left to Right)")
plt.axis('off')
plt.show()

# 4. Compress image by 50% along both axes using np.compress
# First compress rows (y-axis), then columns (x-axis)
rows = np.arange(0, flipped_img.shape[0], 2)  # take every 2nd row
cols = np.arange(0, flipped_img.shape[1], 2)  # take every 2nd col

compressed_img = flipped_img[rows, :][:, cols]

# Show the compressed image
plt.imshow(compressed_img, cmap='gray')
plt.title("Compressed Image (50% both axes)")
plt.axis('off')
plt.show()
```

```python

import numpy as np
a = np.array([[3, 1],[1, 2]])
b = np.array([9, 8])
''' Checking if system of equation has unique solution '''
print(np.linalg.det(a))
# 5.0
''' Since det = 5 which is non-zero. Hence, we have unique solutions
 Finding unique solution '''
print(np.linalg.solve(a, b))
# [ 2.  3.]
''' Calculating Inverse: Since, determinant is non-zero
 hence, matrix is invertible '''
print(np.linalg.inv(a))
# [[ 0.4 -0.2]
#  [-0.2  0.6]]
''' Calculating Rank of the matrix '''
print(np.linalg.matrix_rank(a))
# 2

```

```python
import numpy as np

# Coefficient matrix A
A = np.array([[2, 3, 2],
              [1, 0, 3],
              [2, 2, 3]])

# Right-hand side vector B
B = np.array([1, 2, 3])

# Solve for [x, y, z]
solution = np.linalg.solve(A, B)
print(f"Solution: x = {solution[0]}, y = {solution[1]}, z = {solution[2]}")
```
