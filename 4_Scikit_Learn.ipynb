```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model, datasets
iris = datasets.load_iris()
X = iris.data[:, :2]  # Sepal Width and Sepal Height
Y = iris.target       # Species -> Setosa, Versicolor, Virginica


```

```python
lm = linear_model.LogisticRegression(C=1e5)
lm.fit(X, Y)

```

```python
# Range of X and Y axis
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
# Step size in the mesh
h = .02
# Creating Mesh Grid
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))


```

```python
Z = lm.predict(np.c_[xx.ravel(), yy.ravel()])
# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)
# Labeling
plt.figure(1, figsize=(4, 3))
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks(());
plt.yticks(());
plt.show()

```

```python
plt.scatter(X[:, 0], X[:, 1],c=Y, edgecolors='k', cmap=plt.cm.Paired)


```

Import the 'iris-modified' dataset to python environment. Perform the following operations:



1. Find 3 clusters in this dataset.

2. Find the size of each cluster.  

3. Visualize the clusters using the matplotlib library.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris

# Step 1: Import the 'iris-modified' dataset (for demonstration, we'll use the iris dataset)
# Assuming iris-modified dataset is a CSV file or an already available dataset.
# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Step 2: Apply KMeans clustering to find 3 clusters
kmeans = KMeans(n_clusters=3, random_state=0)
df['Cluster'] = kmeans.fit_predict(df)

# Step 3: Find the size of each cluster
cluster_sizes = df['Cluster'].value_counts()
print(f"Cluster sizes:\n{cluster_sizes}")

# Step 4: Visualize the clusters
plt.figure(figsize=(8, 6))

# Plot the first two features (sepal length and sepal width) for simplicity
plt.scatter(df.iloc[:, 0], df.iloc[:, 1], c=df['Cluster'], cmap='viridis', s=50, alpha=0.7)
plt.title('K-means Clustering (3 clusters)')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.colorbar(label='Cluster')

plt.show()
```
