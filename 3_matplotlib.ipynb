```python
import numpy as np
import matplotlib.pyplot as plt

# Given data
n_grps = np.arange(5)
bar_width = 0.4

men_sc = [20, 30, 10, 50, 90]
err_men_sc = [2, 3, 4, 5, 4]

women = [10, 123, 19, 60, 40]
err_women_sc = [1, 6, 2, 8, 7]

# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plotting Men scores
bars1 = ax.bar(n_grps - bar_width/2, men_sc, bar_width,
               color='skyblue', yerr=err_men_sc,
               capsize=5, label='Men')

# Plotting Women scores
bars2 = ax.bar(n_grps + bar_width/2, women, bar_width,
               color='salmon', yerr=err_women_sc,
               capsize=5, label='Women')

# Add labels and title
ax.set_xlabel('Games')
ax.set_ylabel('Scores')
ax.set_title('Scores of Men and Women Across 5 Games with Observation Errors')
ax.set_xticks(n_grps)
ax.set_xticklabels(['Game 1', 'Game 2', 'Game 3', 'Game 4', 'Game 5'])

# Add legend
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation

n = 30
x = np.arange(0,1,0.001)
y = np.ones( (1000,n) )
for i in range(0,n):
    y[:,i] = np.sin(2 * np.pi * x) * i+1
def func(arg):
     plt.cla()   #Clear axis
     plt.plot(y[:,arg])
     plt.ylim(-30,30)
fig = plt.figure(figsize=(5,4))

to_save = animation.FuncAnimation(fig, func, frames=30)

```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate random data
x, y, z = np.random.rand(3, 50)

# Create a figure and add 3D subplot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the scatter
ax.scatter(x, y, z, c='magenta', marker='^')  # You can change color and marker

# Set axis labels (optional)
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

# Show and rotate the plot
plt.ion()  # Turn on interactive mode
plt.show()

# Rotate the axes and update
for angle in range(0, 360):
    ax.view_init(elev=30, azim=angle)
    plt.draw()
    plt.pause(0.001)

plt.ioff()  # Turn off interactive mode after rotation completes
```

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('pokemon_alopez247.csv')
bulbasaur = list(df.iloc[0,4:11])
charmander = list(df.iloc[3,4:11])
squirtle = list(df.iloc[6,4:11])
pokets = [bulbasaur, charmander, squirtle]
# Converting height from meters to inches
# Weight remains in kg
bul_hw = [df.iloc[0,19]*3.281, df.iloc[0,20]]
char_hw = [df.iloc[3,19]*3.281, df.iloc[3,20]]
squ_hw =  [df.iloc[6,19]*3.281, df.iloc[6,20]]
hw = [bul_hw, char_hw, squ_hw]

```

```python

fig = plt.figure(0)
# Line plot
ax1 = plt.subplot2grid((4,3), (0,0), colspan=3)
ax1.plot(bulbasaur,'g-',bulbasaur,'go')
ax1.plot(charmander,'r-',charmander,'ro')
ax1.plot(squirtle,'b-',squirtle,'bo')
ax1.set_xlim(-0.3,6.3)

```

```python

text_x_coord = [0.045, .19, 0.34, 0.49, 0.64, 0.795, 0.945]
text_y_coord = [0.7, 0.57, 0.65, 0.75, 0.75, 0.75, 0.72]
rot = [0, 45, 45, 45, 45, 45, 45]
txt = [r'$T_{OTAL}$', r'$HP$', r'$A_{TTACK}$', r'$D_{EFENSE}$',
       r'$S_{P}\_A_{TK}$',r'$S_{P}\_D_{EF}$',r'$S_{PEED}$']
for i in range(0, 7):
    ax1.text(text_x_coord[i], text_y_coord[i], txt[i],
             transform = ax1.transAxes,  # makes width and height in percentage
             va = 'top',
             rotation = rot[i],
             bbox = dict(
                     boxstyle = 'round',
                     facecolor = 'wheat',
                     alpha = 0.78))  # alpha -> transparency

```

```python
# continuing from previous code
ax2 = plt.subplot2grid((4,3), (1,0), colspan=1)
ax3 = plt.subplot2grid((4,3), (2,0), colspan=1)
ax4 = plt.subplot2grid((4,3), (3,0), colspan=1)
ax = [ax2, ax3, ax4]
colors = ['g', 'r', 'b']
for i in range(0, 3):
    bp = ax[i].boxplot(pokets[i][1:],patch_artist=True)
    # Adding colors to edges
    for element in ['boxes', 'whiskers', 'fliers', 'means', 'medians', 'caps']:
            plt.setp(bp[element], color='k')
    # Adding color inside the box
    for patch in bp['boxes']:
        patch.set(facecolor=colors[i])
    ax[i].set_ylim(35,70)

```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.0, 5.0, 0.1)
y = np.cos(2 * np.pi * x) * np.exp(-x)

plt.plot(x, y)
plt.title('Damped Cosine Wave')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,1,1,1,2,3,4,4,4,4,3,2,1])
y = np.array([1,2,3,4,4,4,4,3,2,1,1,1,1])

plt.plot(x, y, marker='o', color='blue', linewidth=2)
plt.fill(x, y, color='skyblue', alpha=0.4)  # Optional: fills the shape
plt.title("Custom Polygon Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.axis('equal')  # Makes the x and y axis scale equally
plt.show()
```
